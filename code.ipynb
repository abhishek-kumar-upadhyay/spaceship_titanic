```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("data/train.csv")
```

```python
train
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
train['HomePlanet'].value_counts()
```

```python
train[['Deck', 'Num', 'Side']] = train['Cabin'].str.split('/', expand = True)
train = train.drop(['Cabin'], axis = 1)
train
```

## Treating missing values

```python
train['HomePlanet'].mode()[0]
```

```python
train['HomePlanet'].fillna(train['HomePlanet'].mode()[0], inplace=True)
train['CryoSleep'].fillna(train['CryoSleep'].mode()[0], inplace=True)
train['Destination'].fillna(train['Destination'].mode()[0], inplace=True)
train['VIP'].fillna(train['VIP'].mode()[0], inplace=True)
train['Deck'].fillna(train['Deck'].mode()[0], inplace=True)
train['Num'].fillna(train['Num'].mode()[0], inplace=True)
train['Side'].fillna(train['Side'].mode()[0], inplace=True)

train[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']] = train[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']].fillna(0)

train['Age'].fillna(train['Age'].mean(),inplace=True)
```

```python
train.isnull().sum()
```

```python
train['total_spent'] = train['RoomService'] + train['FoodCourt'] + train['ShoppingMall'] + train['Spa'] + train['VRDeck']
```

```python
train
```

```python
train = train.drop(['PassengerId', 'Name', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck'], axis = 1)
```

```python
train.info()
```

```python
train
```

```python
train['CryoSleep'].replace({True:1,False:0},inplace=True)
train['VIP'].replace({True:1,False:0},inplace=True)
train['Transported'].replace({True:1,False:0},inplace=True)
```

```python
train['Num'] = train['Num'].astype(float)
```

```python
train.info()
```

## Encoding

```python
categorical_cols = [cols for cols in train if train[cols].dtypes=='O']
```

```python
categorical_cols
```

```python
pd.DataFrame(train[categorical_cols].nunique(), columns=['N_unique'])
```

```python
train = pd.get_dummies(train, drop_first=True)
```

```python
train.info()
```

```python
train
```

## Separating Feature and Target Variables

```python
target = train['Transported']
features = train.drop(['Transported'],axis=1)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(features,target,random_state=1,test_size=0.2,stratify=target)
```

## Modelling and Prediction

```python
from sklearn.linear_model import LogisticRegression
logreg_model = LogisticRegression()
logreg_model.fit(xtrain,ytrain)
```

```python
y_pred = logreg_model.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,y_pred))
```

```python
import pickle
pickle.dump(logreg_model, open('models/logreg_model.pkl', 'wb'))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=9, random_state=0)
rf.fit(xtrain,ytrain)
rf_y_pred = rf.predict(xtest)
rf_y_pred
```

```python
print(classification_report(ytest,rf_y_pred))
```

```python
pickle.dump(rf, open('models/rf_model.pkl', 'wb'))
```
